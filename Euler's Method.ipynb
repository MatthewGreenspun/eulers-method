```python
import numpy as np
import pandas as pd
from math import sin, cos, e, pi
import matplotlib.pyplot as plt
```

```python
def euler(dy_dx, x0, y0, xn, delta_x): 
    x_arr = []
    y_arr = [y0]
    y_prime_arr = []
    next_arr = []
        
    for i in np.arange(x0, xn, delta_x):
        x_arr.append(i)
        y_prime = dy_dx(x_arr[-1], y_arr[-1])
        y_prime_arr.append(y_prime)
        next = y_arr[-1] + y_prime * delta_x
        next_arr.append(next)
        if i < xn-delta_x:
            y_arr.append(next)
    
    plt.plot(x_arr, y_arr)
    plt.scatter(x_arr, y_arr)
    
    table = pd.DataFrame({
        "x": x_arr,
        "y": y_arr,
        "y'": y_prime_arr,
        "y + y' * dx": next_arr
    })
    return table
```

```python
# 8
def f(x, y): 
    return x - y**2
euler(f, x0=0, y0=1, xn=2, delta_x=0.25)
```

```python
# 9
def f(x, y):
    return cos(y)
euler(f, x0=0, y0=1, xn=2, delta_x=0.5)
```

```python
# 10 
def f(x, y):
    return e**(-y)
table = euler(f, x0=0, y0=0, xn=1, delta_x=0.1)
table
```

```python
# 11
def f(x, y):
    return sin(pi * x)
euler(f, x0=0, y0=0, xn=1, delta_x=0.2)
```
